```python
from pathlib import Path

#Creating a folder to save subtitles
subs_dir = Path("/content/drive/My Drive/ProjectSemester1")
subs_dir.mkdir(parents=True, exist_ok=True)
```

```python
#Download yt-dlp
#This tool will allow me to obtain the desired data from YouTube.
!pip install yt-dlp
```

```python
output_template = subs_dir / "%(title)s.%(ext)s"
```

```python
import json
import subprocess
#In this section I will get the YouTube subtitles using yt-dlp

# Get the list of videos appearing in the playlist
youtube_playlist_url = "https://www.youtube.com/playlist?list=PLAttudMji90ffG7JlmJ2t0zE-CUtRlnib"
yt_dlp_result = subprocess.run(["yt-dlp", "--flat-playlist", "-J", youtube_playlist_url], capture_output=True, text=True)
#Converts the output from JSON to a dictionary
playlist_info = json.loads(yt_dlp_result.stdout)
#List of playlist video IDs
videos_ids = [entry['id'] for entry in playlist_info['entries']]
#Check that all the videos found
print(f"Found {len(videos_ids)} videos in the playlist.")

#Downloading the subtitles
for vid in videos_ids:
    youtube_playlist_url = f"https://www.youtube.com/watch?v={vid}"
    print(f"Downloading subtitles for: {youtube_playlist_url}")
    yt_dlp_result = subprocess.run([
    "yt-dlp",
    "--write-subs",
    "--sub-lang", "iw",
    "--convert-subs", "srt",
    "--skip-download",
    "-o", output_template,
    youtube_playlist_url
    ], capture_output=True, text=True)

    print(yt_dlp_result.stderr)
    #Printing the error in case it happens
```
