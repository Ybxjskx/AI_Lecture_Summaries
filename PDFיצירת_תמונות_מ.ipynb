```python
from pdf2image import convert_from_path
import os
#I will convert the presentation into images of the pages
#The purpose of this is to make it possible to compare the video with
# the presentation pages.

#Path to the pdf lecture
pdf_lecture_path = "/content/drive/MyDrive/ProjectSemester1/PresentationsPDF/Lecture 8a.pdf"

output_folder = "/content/drive/MyDrive/ProjectSemester1/slides"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

slides = convert_from_path(pdf_lecture_path, dpi=300) #Convert presentation pages to images for use
for i, slide in enumerate(slides):
    slide.save(f"{output_folder}/slide_{i+1}.png", "PNG") #saving

print("The save was successful")
```
