```python
from pptx import Presentation
from pathlib import Path
import os

PPT_PATH = "/content/drive/MyDrive/ProjectSemester1/PresentationPPTM/Lecture 8a.pptm"

NOTES_DIRECTORY_PATH = Path("/content/drive/MyDrive/ProjectSemester1/summaries")
PPT_OUTPUT_PATH = "/content/drive/My Drive/ProjectSemester1/PresentationFinal/presentation_with_notes.pptx"

def insert_summeries_to_presentation(ppt_path: Path, notes_dir: Path, output_path: Path) -> None:
  presentation = Presentation(ppt_path)
  for i, slide in enumerate(presentation.slides):
      summery_file_path = notes_dir / f"summary_slide_{i}.txt"

      with summery_file_path.open("r", encoding="utf-8") as f:
          note_text = f.read()

      notes_slide = slide.notes_slide
      text_frame = notes_slide.notes_text_frame
      text_frame.clear()
      text_frame.text = note_text

  output_dir = os.path.dirname(output_path)
  if not os.path.exists(output_dir):
      os.makedirs(output_dir)

  presentation.save(output_path)

insert_summeries_to_presentation(PPT_PATH, NOTES_DIRECTORY_PATH, PPT_OUTPUT_PATH)
```
